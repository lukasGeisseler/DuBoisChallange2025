```python
import pandas as pd

from matplotlib import font_manager
import matplotlib.pyplot as plt

for fontpath in font_manager.findSystemFonts(fontpaths=None, fontext="ttf"):
    if "Cobol".lower() in fontpath.lower():
        print(fontpath)

cobol_regular = font_manager.FontProperties(
    fname="/Users/lukasgeisseler/Library/Fonts/cobol.ttf"
)
cobol_bold = font_manager.FontProperties(
    fname="/Users/lukasgeisseler/Library/Fonts/Cobol W00 Bold.ttf"
)

%matplotlib inline
%config InlineBackend.figure_format='retina'

df_plate_19 = pd.read_csv("./data.csv")
df_plate_19 = df_plate_19.sort_values(by="Date", ascending=False, ignore_index=True)
df_plate_19 = df_plate_19.reset_index()
df_plate_19
```

```python
fig, ax = plt.subplots(figsize=(8, 10), facecolor="#DDCFBE")

# Draw Barplot
ax.barh(
    y="index",
    width="Land",
    height=0.5,
    data=df_plate_19,
    color="#D03D4A",
    edgecolor="#52231D",
    linewidth=0.4,
)

# Add the bartext for first an last entry
for i in range(len(df_plate_19)):
    if i in [0, 24]:
        ax.text(
            df_plate_19.Land[i] / 2,
            i - 0.05,
            f"{df_plate_19.Land[i]:,.0f}",
            fontsize=11,
            color="#1D1E20",
            ha="center",
            va="center",
            font=cobol_bold,
        )

# Adjust appearance
ax.set_ylim(
    bottom=-1.5,
    top=24.25,
)
ax.set_yticks(df_plate_19.index)
ax.set_yticklabels(
    labels=df_plate_19.Date,
    font=cobol_regular,
    fontsize=13,
    horizontalalignment="left",
)
ax.set_xticks([])
ax.set_ylabel(" ", font=cobol_regular, fontsize=13)
ax.tick_params(tick1On=False, pad=25)
ax.set(frame_on=False)

# add Title
ax.set_title(
    "Acres of Land owned by\n black georgians.".upper(),
    color="#1D1E20",
    font=cobol_bold,
    fontweight="bold",
    fontsize=15,
)


plt.show()
```
